# Feature Engineering

```python
covid.head()
```

```python
covid.info()
```

```python
age = covid['age']
```

```python
def age(x):
  if x>=0 and x<=14:
    return 'Children'
  elif x>=15 and x<=24:
    return 'Young'
  elif x>=25 and x<=64:
    return 'Adult'
  else:
    return 'Old'
```

```python
covid['age'] = covid['age'].apply(age)
```
