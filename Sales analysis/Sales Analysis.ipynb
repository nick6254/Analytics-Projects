# Outlier Treatment

```python
# Outlier treatment

plt.figure(figsize=(10,12))

c=1
for i in num_col:
  plt.subplot(4,2,c)
  sns.boxplot(x =sales[i])
  plt.title(i)
  c+=1
plt.tight_layout()
plt.show()
```

```python
for i in num_col:
  # q1,q3 = np.quantile(sales[i],[0.25,0.75])
  # iqr = q3-q1
  # ul = q3+(1.5*iqr)
  # ll = q1-(1.5*iqr)
  ll,ul = np.quantile(sales[i],[0.10,0.95])
  sales.loc[sales[i]<ll,i] = ll
  sales.loc[sales[i]>ul,i] = ul
```

```python
plt.figure(figsize=(10,12))

c=1
for i in num_col:
  plt.subplot(4,2,c)
  sns.boxplot(x =sales[i])
  plt.title(i)
  c+=1
plt.tight_layout()
plt.show()
```
